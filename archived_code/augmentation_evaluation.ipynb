```python
import pandas as pd
import ast
```

```python
with open('./re.txt', 'r') as f:
    lines = f.readlines()
```

```python
stat = []
level2 = ['AM', 'AN', 'AS', 'AZ', 'HM', 'HN', 'HS', 'OC', 'OM', 'ON', 'OS', 'ZI', 'ZN']
level3 = ['ON_R', 'ON_S', 'ON_R2', 'HM_S']
level4 = ['ON_RG', 'ON_SR', 'ON_RG2', 'HM_SR', 'ON_SR2', 'ON_RGn', 'ON_RSn', 'ON_RNn', 'ON_RDn']
level5 = ['ON_SRG', 'HM_SRG']
levels = [level2, level3, level4, level5]

for idx, l in enumerate(lines):
    if len(l) == 1:  # empty line
        continue
    if l[0] == '[':  # data line
        item = ast.literal_eval(l)
        new_item = [item[1], item[2]['loss']]
        key = lines[idx-1].split('-')[0]
        is_lumen = 'Lumen' in lines[idx-1]
        new_item.append('Lumen' if is_lumen else 'Media')
        new_item.append(key)
        for idx, lvl in enumerate(levels):
            if key in lvl:
                new_item.append(idx + 2)
        stat.append(new_item)
```

```python
df = pd.DataFrame(stat, columns=['test_loss', 'loss', 'target', 'augmentation', 'level'])
df.head()
```

```python
ds = df.groupby(['level', 'target', 'augmentation']).describe()['test_loss']
ds.to_excel('stat.xls')
ds
```

```python
def sort_lists(listx, listy):
    xy = list(zip(listx, listy))
    xy = sorted(xy, key=lambda x:(x[0][0], x[-1]))
    return xy
```

```python
sort_lists(list(ds.index), list(ds['mean']))
```
