```python
import pandas as pd
import ast
```

```python
with open('./result.txt', 'r') as f:
    lines = f.readlines()
```

```python
stat = []

for idx, l in enumerate(lines):
    if len(l) == 1:  # empty line
        continue
    if l[0] == '[':  # data line
        item = ast.literal_eval(l)
        new_item = [item[1], item[2]['loss']]
        key = lines[idx-1][:2]
        is_lumen = 'Lumen' in lines[idx-1]
        new_item.append('Lumen' if is_lumen else 'Media')
        new_item.append(key)
        stat.append(new_item)
```

```python
df = pd.DataFrame(stat, columns=['test_loss', 'loss', 'target', 'augmentation'])
df.head()
```

```python
ds = df.groupby(['target', 'augmentation']).describe()['test_loss']
ds
```

```python
def sort_lists(listx, listy):
    xy = list(zip(listx, listy))
    xy = sorted(xy, key=lambda x:(x[0][0], x[-1]))
    final_x, final_y = zip(*xy)
    return xy
```

```python
sort_lists(list(ds.index), list(ds['mean']))
```
